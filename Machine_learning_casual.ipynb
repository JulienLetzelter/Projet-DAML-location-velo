# Machine Learning - Zoom sur la prédiction du nombre de locations occasionnelles de vélos

Ce notebook implémente des modèles de machine learning pour prédire le nombre de locations occasionnelles de vélos en utilisant les données du dataset bike sharing. Suite à la première analyse des données et au travail qui a été fait sur toute la base de donnée, il semble pertinent de distinguer les cas "casual" et "registered" qui ne répondent pas au même besoins et au mêmes habitudes d'utilisation. Nous nous attendons donc à des prédictions plus efficaces de la part des algorihtmes.
